```python
secret: int = 5
guess: int = 7

if guess == secret:
    print('just right')
elif guess < secret:
    print('too low')
else:
    print('too high')
```

4.2


```python
small: bool = True
green: bool = False

if small and green :
    print('pea')
elif small and not green:
    print('cherry')
elif not small and green:
    print('watermelon')
else:
    print('pumpkin')
```

```python
guess_me: int = 7
number: int = 1

while True:
    if number < guess_me:
        print('too low')
    elif number > guess_me:
        print('too high')
    else:
        print('found it!')
        break
   
    number += 1
```

```python
guess_me: int = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number > guess_me:
        print('too high')
    else:
        print('found it!')
        break
```
