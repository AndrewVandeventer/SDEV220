Write a Python app that has the following classes:
1. A super class called Vehicle, which contains an attribute for vehicle type, such as car, truck, plane, boat, or a broomstick.
2. A class called Automobile which will inherit the attributes from Vehicle and also contain the following attributes:
year
make
model
doors (2 or 4)
roof (solid or sun roof).

Write an app that will accept user input for a car. The app will store "car" into the vehicle type in your Vehicle super class. 

The app will then ask the user for the year, make, model, doors, and type of roof and store the data in the attributes above.

The app will then output the data in an easy-to-read and understandable format, such as this:
  Vehicle type: car
  Year: 2022
  Make: Toyota
  Model: Corolla
  Number of doors: 4
  Type of roof: sun roof

```python
class Vehicle:
    def __init__(self, vic_typ):
        """Initialize a vehicle with a type."""
        self.vic_typ = vic_typ
        
class Car(Vehicle):
    def __init__(self, vic_typ, year, make, model, door_num, roof):
        """Initialize a car with specific attributes."""
        super().__init__(vic_typ)
        self.year = year
        self.make = make
        self.model = model
        self.door_num = door_num
        self.roof = roof
        
    def display_info(self):
        """Display the car's information."""
        print("Vehicle Information:")
        print(f"Type: {self.vic_typ}")
        print(f"Year: {self.year}")
        print(f"Make: {self.make}")
        print(f"Model: {self.model}")
        print(f"Number of Doors: {self.door_num}")
        print(f"Roof Type: {self.roof}")

# Validate user input
while True:
    try:
        year = int(input("Enter vehicle year: "))
        if 1886 <= year <= 2026:
            break
        else:
            print("Invalid year. Please enter a valid year between 1886 and 2026.")
    except ValueError:
        print("Please enter a number.")

make = input("Enter make: ")
model = input("Enter model: ")

while True:
    try:
        door_num = int(input("Enter number of doors (2 or 4): "))
        if door_num in [2, 4]:
            break
        else:
            print("Invalid input. Please enter 2 or 4.")
    except ValueError:
        print("Please enter a number.")

while True:
    roof = input("Enter roof type (sunroof or solid): ").lower()
    if roof in ["sunroof", "solid"]:
        break
    else:
        print("Invalid input. Please enter 'sunroof' or 'solid'.")

# Create the car
car = Car(
    vic_typ="car",
    year=year,
    make=make,
    model=model,
    door_num=door_num,
    roof=roof
)

# Display info
car.display_info()
```
